```python
import pandas as pd

df = pd.read_csv('autos_infracao.csv',sep=';')
df
```

```python
df['data_hora_infracao'] = pd.to_datetime(df['data_hora_infracao'])

df['ano'] = df['data_hora_infracao'].dt.year

df_2023 = df[df['ano'] == 2023]

contagem_bairros = df_2023['bairro'].value_counts().head(10)

print('Os 10 bairros com mais infrações em 2023:')
print(contagem_bairros)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
contagem_bairros.plot(kind='bar', color='skyblue')
plt.xlabel('Bairros')
plt.ylabel('infracoes')
plt.title('10 Bairros com Mais Infrações Sanitarias em 2023')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
